```python
import pandas as pd
data = pd.read_csv("movie.csv",encoding = 'utf_8_sig')
data.head()
```

```python
row,col = data.shape
```

```python
data['唯一分類'] = "create"
for i in range(0,row):
    if data['分類'][i][4:5] == "\'":
        movie_type = data['分類'][i][2:4]
    elif data['分類'][i][4:5] == "/":
        movie_type = data['分類'][i][2:4]
    else:
        movie_type = data['分類'][i][2:5]
    data.loc[i,'唯一分類'] = movie_type
data
```

```python
data.to_csv('movie_new.csv', encoding='utf_8_sig' , index = False)
```

```python
from sklearn.model_selection import train_test_split

train , test = train_test_split(data, random_state = 103,test_size=0.2)
```

```python

train["唯一分類"].value_counts().sort_index().plot( kind = "barh", title = "訓練集分類")

test["唯一分類"].value_counts().sort_index().plot( kind = "barh", title = "測試集分類")
```

```python
import jieba
import re

def content_cut(text):
    specialChars = "!#$%^&*()《  》"
    text = text.replace("★" , "").replace("\r\n" , "").replace("！","").replace("《","").replace("》","").replace("，","").replace("/","").replace("（","").replace("）","").replace(specialChars,"")
    return " ".join(list(jieba.cut(text)))

train['分詞劇情'] = train['劇情介紹'].map(content_cut)
test['分詞劇情'] = test['劇情介紹'].map(content_cut)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(sublinear_tf=False, stop_words=None, token_pattern="(?u)\\b\\w+\\b", smooth_idf=True, norm='l2')
X_train = vectorizer.fit_transform(train["分詞劇情"])
X_test = vectorizer.transform(test["分詞劇情"])
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=11,weights="distance")
knn.fit(X_train, train["唯一分類"])
```

```python
Y_test = knn.predict(X_test)
```

```python
socre=knn.score(X_test,test["唯一分類"])
print('精準度:{:2f}'.format(socre))
```
